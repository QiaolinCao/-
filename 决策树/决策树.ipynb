西瓜数据集2.0

```python
df =pd.read_csv("./watermelon_2.0.csv", index_col="编号")
df  = df [["脐部", "色泽", "根蒂", "敲声", "纹理", "触感", "好瓜"]]
df
```

```python
mapping = {}
for col_name in df.columns:
    df[col_name], mapping[col_name] = pd.factorize(df[col_name])
```

```python
df
```

```python
mapping
```

```python
test_set = df.loc[[4,5,8,9,11,12,13]]
train_set = df.loc[df.index.difference(test_set.index)]
```

运用数据集，生成基于信息增益准则，不剪枝的决策数  
所有属性均为离散变量  
类别标签为“好瓜”  
对应周志华P81页图4.5

```python
divide_method = "information_gain"
pruning_method = "None"
search_method = "depth_first"
tree = DecisionTree(divide_method, pruning_method, search_method)
structure = DataStructure([], list(df.columns[:-1]), "好瓜")
tree.fit(train_set, test_set, structure)
```
